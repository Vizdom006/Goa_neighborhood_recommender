## Data Sources

The major Data sources for our project are

 - **Makaan.com**: A real estate brokerage website from which we will scrape the data from.
 - **Foursquare API** : We will get all the venue's data in the neighbourhood using this API
 - **Nominatim API** : To get Lat and long co-ordinates for our neighbourhoods.

### Data from Makaan

A table containing
- Locality
- Average rent in each locality

### Data from Foursquare

We pass the Neighbourhood name to get the list of all possible venues available in the Neighbourhood, Based on the number of restaurants we do a competitive analysis and rank the neighbourhoods accordingly.
