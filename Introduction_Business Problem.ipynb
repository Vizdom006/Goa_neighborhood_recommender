## Problem Statement

The project aims to build a simple recommending tool that suggests neighbourhoods in goa for anyone based on their preferences.

we will be using features like average rent in each neighbourhood, amneties available in each neighbourhood as features while in modelling we use the K-means clustering algorithm (Unsupervised algorithm) to cluster our neighbourhoods based on similarity and give recomendation for consumers to select a suitable neighbourhood to stay in.


## The Location

We will be using around 60 neighbourhoods in **Goa**

## Modelling approach

 We use **K means clustering** algorithm for clustering similar neighbourhoods.
